```python
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, AgentType, Tool

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

def search_tool(query):
    return f"You searched for: {query}"

tools = [
    Tool(
        name="SearchTool",
        func=search_tool,
        description="Useful for answering search-related questions"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

result = agent.invoke("Search for Generative AI")
print(result["output"])
```

```python
from langchain.agents import create_agent 
def get_weather(city:str)->str:
    """ Get the weather for a city"""
    return f"The weather in {city} is raining."

    
agent = create_agent(
    model = "gpt-5",
    tools =[get_weather],
    system_prompt = "You are a helpful agent"
)
agent
```

```python
### Run the agent
response=agent.invoke({"messages":[{"role":"user", "content":"what the weather in Dallas "}]})
```

```python
response["messages"][-1].content
```

```python
agent.invoke({"messages":"What is the weather in Hyberabad in india"})
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    "gpt-4o-mini",
    model_provider="openai"
)

response = model.invoke("write me 200 words paragraph about Nellore")

print(response.content)
```

```python
for chunk in model.stream("why usa is richest country"):
    print(chunk.text, end="", flush=True)

```

```python
responses= model.batch([
    "what is malikaja in nellore",
    "what is the special in nellore",
    "what is scope of amaravathi"
])
for response in responses:
    print(response)
```

```python
model.batch([
    "why nellore is famous",
    "what is use of aiport",
    "what are causes for data centre"
],
config={
    'max_concurrency':5
}
)
```
