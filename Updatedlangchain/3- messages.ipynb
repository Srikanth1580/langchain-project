```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]= os.getenv("GROQ_API_KEY")

model = init_chat_model(
    "llama-3.3-70b-versatile",
    model_provider="groq"
)
```

## Message Prompts 

#### Altenatively , you can can pass a list of messages to the model by provideing the list of objects.
#### Message Types
###### System Message
###### Human Message
###### AI Message
###### Tool Message

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
messsages=[
    SystemMessage("You are an expert "),
    HumanMessage("write a plociy for politcal startagey")
]
response = model.invoke(messsages)
print(response.content)
```

```python
messsages=[
    SystemMessage("You are an expert "),
    HumanMessage("write a Rest API startagey")
]
response = model.invoke(messsages)
print(response.content)
```

```python

SYS= SystemMessage("You are an expert ")
messsages=[
    SYS,
    HumanMessage("write a Rest API startagey")
]
response = model.invoke(messsages)
print(response.content)
```

```python
### Detailed info to the LLM through the system message 

System_message = SystemMessage("""
You are an IAS Officer with good knowledege in Indian education system.
You need to provide good stratagies to implement the good practices to the drop out students.
Can you please provide me the real time examples how can we able to reduce the drop out students.
""")
messsages =[
    System_message,
    HumanMessage("How to reduce drop out students from school")
]
response = model.invoke(messsages)
print(response.content)
```

```python
from langchain.messages import AIMessage , SystemMessage, HumanMessage

ai_msg = AIMessage("I AM VERY HAPPY TO GIVE POLITICAL STRATGIES")

messsages= [
    SystemMessage("You are an Helpful political stratagist"),
    HumanMessage("Give me political stratagies for my village panchayat elections"),
    ai_msg,
    HumanMessage("Give me real time things")

]
response = model.invoke(messsages)
print(response.content)
```

```python
from langchain.messages import AIMessage
from langchain.messages import ToolMessage

# After a model makes a tool call
# (Here, we demonstrate manually creating the messages for brevity)
ai_message = AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_weather",
        "args": {"location": "San Francisco"},
        "id": "call_123"
    }]
)

# Execute tool and create result message
weather_result = "Sunny, 72°F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # Must match the call ID
)

# Continue conversation
messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  # Model's tool call
    tool_message,  # Tool execution result
]
response = model.invoke(messages)  # Model processes the result
```
