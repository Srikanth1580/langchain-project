```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("What is famous for nellore")
response
```

```python
from langchain.tools import tool

@tool
def get_data (location:str)->str:
    """ Get the data from particular place """
    return f"It is famous {location} for malikaja"


model_with_tools = model.bind_tools([get_data])
```

```python
response = model_with_tools.invoke("what the famous for Dallas")
print(response)

for tool_call in response.tool_calls:

    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```

### Tool Execution Loops

```python
messages =[{"role":"user","content":"what is the famous for noellore"}]
ai_mssg = model_with_tools.invoke(messages)
messages.append(ai_mssg)

for tool_call in ai_mssg.tool_calls:
    tool_result= get_data.invoke(tool_call)
    messages.append(tool_result)

final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
